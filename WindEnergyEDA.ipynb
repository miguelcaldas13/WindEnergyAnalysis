### Setting up Notebook with the necessary tools

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
location1 = pd.read_csv("./Location1.csv")
location2 = pd.read_csv("./Location2.csv")
location3 = pd.read_csv("./Location3.csv")
location4 = pd.read_csv("./Location4.csv")
```

```python
print("Information of Location1:")
location1.info()

print("\nInformation of Location2:")
location2.info()

print("\nInformation of Location3:")
location3.info()

print("\nInformation of Location4:")
location4.info()
```

```python
from IPython.display import display
print("First Five Rows of Location1:")
display(location1.head())

print("\nFirst Five Rows of Location2:")
display(location2.head())

print("\nFirst Five Rows of Location3:")
display(location3.head())

print("\nFirst Five Rows of Location4:")
display(location4.head())
```

```python
#Check for missing values in each DataFrame
null_values_location_1 = location1.isna().sum()
null_values_location_2 = location2.isna().sum()
null_values_location_3 = location3.isna().sum()
null_values_location_4 = location4.isna().sum()

print("Null values in location1:\n", null_values_location_1)
print("\nNull values in location2:\n", null_values_location_2)
print("\nNull values in location3:\n", null_values_location_3)
print("\nNull values in location4:\n", null_values_location_4)
```

```python
#Check for duplicated rows in each Dataframe
duplicated_rows1 = location1.duplicated().sum()
duplicated_rows2 = location2.duplicated().sum()
duplicated_rows3 = location3.duplicated().sum()
duplicated_rows4 = location4.duplicated().sum()

print("Total of duplicated rows in location1:\n", duplicated_rows1)
print("\nTotal of duplicated rows location2:\n", duplicated_rows2)
print("\nTotal of duplicated rows in location3:\n", duplicated_rows3)
print("\nTotal of duplicated rows in location4:\n", duplicated_rows4)
```

```python
#Conver 'Time' column to datetime dtype in each Dataframe
location1['Time'] = pd.to_datetime(location1['Time'])
location2['Time'] = pd.to_datetime(location2['Time'])
location3['Time'] = pd.to_datetime(location3['Time'])
location4['Time'] = pd.to_datetime(location4['Time'])

print("Updated Dataframe Location1:")
location1.info()
print("\nUpdated Dataframe Location2:")
location2.info()
print("Updated Dataframe Location3:")
location3.info()
print("\nUpdated Dataframe Location4:")
location4.info()
```

```python
statistics_location1 = location1.describe()
statistics_location2 = location2.describe()
statistics_location3 = location3.describe()
statistics_location4 = location4.describe()

# Displaying the summary statistics
print("Statistics for Location1 DataFrame:\n", statistics_location1)
print("\nStatistics for Location2 DataFrame:\n", statistics_location2)
print("\nStatistics for Location3 DataFrame:\n", statistics_location3)
print("\nStatistics for Location4 DataFrame:\n", statistics_location2)
```

```python
location1_correlation = location1.corr()
location2_correlation = location2.corr()
location3_correlation = location3.corr()
location4_correlation = location4.corr()

# Displaying the summary statistics
print("Data correlation for Location1 DataFrame:\n", location1_correlation)
print("\nData correlation for Location2 DataFrame:\n", location2_correlation)
print("\nData correlation for Location3 DataFrame:\n", location3_correlation)
print("\nData correlation for Location4 DataFrame:\n", location4_correlation)
```

```python
print("Displaying heatmap of Location1: ")
plt.figure(figsize=(15,8))
sns.heatmap(location1_correlation, annot=True, cmap="YlGnBu")
```

```python
print("Displaying heatmap of Location2: ")
plt.figure(figsize=(15,8))
sns.heatmap(location2_correlation, annot=True, cmap="YlGnBu")
```

```python
print("Displaying heatmap of Location3: ")
plt.figure(figsize=(15,8))
sns.heatmap(location3_correlation, annot=True, cmap="YlGnBu")
```

```python
print("Displaying heatmap of Location4: ")
plt.figure(figsize=(15,8))
sns.heatmap(location4_correlation, annot=True, cmap="YlGnBu")
```

```python
#Energy production comparison between Locations
sumof_power_location1 = location1['Power'].sum()
sumof_power_location2 = location2['Power'].sum()
sumof_power_location3 = location3['Power'].sum()
sumof_power_location4 = location4['Power'].sum()

power_data = {'Location': ['Location1','Location2','Location3','Location4'],
              'Total Power Output': [sumof_power_location1, sumof_power_location2, sumof_power_location3, sumof_power_location4]}

power_data_df = pd.DataFrame(power_data)
print(power_data_df)
```
